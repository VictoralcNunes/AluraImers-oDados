####Solução Desafio 1:

```
mean=dados_sem_notas_zero['NU_NOTA_TOTAL'].mean()
median=dados_sem_notas_zero['NU_NOTA_TOTAL'].median()
mode=dados_sem_notas_zero['NU_NOTA_TOTAL'].mode()[0]
```

```
mean
```

```
median
```

```
mode
```

```
plt.figure(figsize=(12,8))
sns.histplot(dados_sem_notas_zero, x = "NU_NOTA_TOTAL", hue="Q025", kde=True, stat="probability")
plt.axvline(mean, color='r', linestyle='--')
plt.axvline(median, color='g', linestyle='-')
plt.axvline(mode, color='b', linestyle='-')
```

####Solução Desafio 2:

```
import numpy as np
```

```
mask = np.triu(np.ones_like(correlacao, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True,)
sns.heatmap(correlacao, mask=mask, cmap=cmap, center=0, annot=True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
